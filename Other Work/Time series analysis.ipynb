```python
data_gr = pd.read_csv(
    "oxford-government-response.csv",
    keep_default_na=False,
    na_values=[""],
)

data_epi = pd.read_csv(
    "epidemiology.csv",
    keep_default_na=False,
    na_values=[""],
)

data_demo = pd.read_csv(
    "demographics.csv",
    keep_default_na=False,
    na_values=[""],
)
```

```python
data_gr = data_gr[(data_gr['date'] > '2020-01-24')]
data_epi = data_epi[(data_epi['date'] < '2020-10-30')]

region1_gr = data_gr[(data_gr.key == 'US_CA')]
region1_gr = region1_gr.reset_index(drop=True)
region1_epi = data_epi[(data_epi.key == 'US_CA')]
region1_epi = region1_epi.reset_index(drop=True)

region2_gr = data_gr[(data_gr.key == 'US_MI')]
region2_gr = region2_gr.reset_index(drop=True)
region2_epi = data_epi[(data_epi.key == 'US_MI')]
region2_epi = region2_epi.reset_index(drop=True)

# region3_gr = data_gr[(data_gr.key == 'US_NY')]
# region3_gr = region3_gr.reset_index(drop=True)
# region3_epi = data_epi[(data_epi.key == 'US_NY')]
# region3_epi = region3_epi.reset_index(drop=True)
                    
print(len(region1_gr))
print(len(region1_epi))
print(len(region2_gr))
print(len(region2_epi))
# print(len(region3_gr))
# print(len(region3_epi))
display(region1_gr)
display(region1_epi)
```

```python
plt.figure(figsize=(12,5))
plt.xlabel('')
ax1 = region1_epi.new_confirmed.plot(color='blue', grid=True, label='CA cases')
ax2 = region1_gr.stringency_index.plot(color='red', grid=True, secondary_y=True, label='CA S index')
ax1 = region2_epi.new_confirmed.plot(color='green', grid=True, label='MI cases')
ax2 = region2_gr.stringency_index.plot(color='orange', grid=True, secondary_y=True, label='MI S index')
# ax1 = region3_epi.new_confirmed.plot(color='purple', grid=True, label='NY cases')
# ax2 = region3_gr.stringency_index.plot(color='teal', grid=True, secondary_y=True, label='NY S index')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
```
